### 2.1.3. Markerì™€ popup í™œìš©

```python
# â“ ì„  ê·¸ë¦¬ê¸°
m = folium.Map([37.51434733724219, 127.07303593988632], zoom_start=15)

line_coords = [
    (37.51108512059325,127.07397154356342),
    (37.51216756170485,127.07197060104319),
    (37.51587923093469,127.07273202617604)
]

folium.PolyLine(
    locations=line_coords,
    color='blue',  # ì„  ìƒ‰ìƒ
    weight=5,  # ì„  êµµê¸°
    opacity=0.7,  # ì„  íˆ¬ëª…ë„
    tooltip="ì¢…í•©ìš´ë™ì¥ì—­ì—ì„œ ì ì‹¤ì•¼êµ¬ì¥ ì°ê³ , ì˜¬ë¦¼í”½ì£¼ê²½ê¸°ì¥ìœ¼ë¡œ"
).add_to(m)
m
```

```python
# â“‘ ì› ê·¸ë¦¬ê¸°
m = folium.Map([37.51434733724219, 127.07303593988632], zoom_start=15)

popup_content = "ì ì‹¤ìƒˆë‚´ìˆ˜ì˜ì¥"
popup = folium.Popup(popup_content, max_width=300)

folium.CircleMarker(
    location=[37.514507502409714, 127.0762258398054],
    radius=15, # ì› í¬ê¸°
    fill_color="yellow",# ì±„ìš°ê¸° ìƒ‰ìƒ
    fill_opacity=0.6, # ì±„ìš°ê¸° íˆ¬ëª…ë„
    color="orange", # í…Œë‘ë¦¬ ìƒ‰ìƒ
    weight=3, # í…Œë‘ë¦¬ ë‘ê»˜
    opacity=1, # í…Œë‘ë¦¬ íˆ¬ëª…ë„
    popup=popup
).add_to(m)
m
```

```python
# â“’ ì‚¬ê°í˜• ê·¸ë¦¬ê¸°
m = folium.Map([37.51434733724219, 127.07303593988632], zoom_start=15)

rectangle_coords = [[37.5174,127.0703], [37.5121,127.0779]]

# ì‚¬ê°í˜• ê·¸ë¦¬ê¸°
folium.Rectangle(
    bounds=rectangle_coords, #ì‚¬ê°í˜• ì¢Œí‘œ ì„¤ì • (ì‚¬ê°í˜•ì˜ ëª¨ì„œë¦¬ ì¢Œí‘œ ì…ë ¥)
    fill_color='skyblue',  # ì±„ìš°ê¸° ìƒ‰ìƒ
    fill_opacity=0.4,  # ì±„ìš°ê¸° íˆ¬ëª…ë„
    color='blue',  # í…Œë‘ë¦¬ ìƒ‰ìƒ
    weight=2,  # í…Œë‘ë¦¬ ë‘ê»˜
    opacity=0.7  # íˆ¬ëª…ë„
).add_to(m)

m
```

```python
# â““ ë‹¤ê°í˜• ê·¸ë¦¬ê¸°
m = folium.Map([37.51434733724219, 127.07303593988632], zoom_start=15)

# polygonì„ ê·¸ë ¤ë³´ì
locations = [
    [37.51108512059325,127.07397154356342],
    [37.51216756170485,127.07197060104319],
    [37.51587923093469,127.07273202617604]
]

popup_content = "ì¢…í•©ìš´ë™ì¥-ì ì‹¤ì•¼êµ¬ì¥-ì˜¬ë¦¼í”½ì£¼ê²½ê¸°ì¥"
popup = folium.Popup(popup_content, max_width=300)

folium.Polygon(
    locations=locations,
    fill_color="skyblue",
    fill_opacity=0.5,
    color="blue",
    weight=3, # í…Œë‘ë¦¬ ë‘ê»˜
    opacity=1, # í…Œë‘ë¦¬ íˆ¬ëª…ë„
    popup=popup
).add_to(m)


m
```

### 2.1.5. layer controlê³¼ FeatureGroup í™œìš©

```python
m = folium.Map((37.51434733724219, 127.07303593988632), zoom_start=15)

group_1 = folium.FeatureGroup("ì ì‹¤ì•¼êµ¬ì¥").add_to(m)
folium.Marker((37.51216756170485,127.07197060104319), icon=folium.Icon(icon="heart")).add_to(group_1)

group_2 = folium.FeatureGroup("ì˜¬ë¦¼í”½ì£¼ê²½ê¸°ì¥").add_to(m)
folium.Marker((37.51587923093469,127.07273202617604), icon=folium.Icon("green")).add_to(group_2)

#collapsed=Falseë¥¼ í•´ë‘ë©´ í•­ìƒ í¼ì³ë‘˜ ìˆ˜ ìˆê³ , ê¸°ë³¸ê°’ì€ True
folium.LayerControl(collapsed=False).add_to(m)

m
```
